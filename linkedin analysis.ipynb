```python
for data in l_Data1:
    if data is not None:
        split=data.split(" Â· ")
        l_Company.append(split[0])
        l_Location.append(split[1])
        l_Applicants_Data.append(split[3])
    else:
        l_Company.append(None)
        l_City.append(None)
        l_State.append(None)
        l_Country.append(None)
        l_Applicants.append(None)
    
for location in l_Location:
    split=location.split(", ")
    if(len(split)==1 and split[0] != "India"):
        l_City.append(split[0])
        l_State.append(None)
        l_Country.append(None)
        
    elif(len(split)==1) and split[0]=="India":
        l_City.append(None)
        l_State.append(None)
        l_Country.append(split[0])
        
    elif(len(split)==2):
        if(split[1]=="India"):
            l_City.append(None)
            l_State.append(split[0])
            l_Country.append(split[1])
            
        else:
            l_City.apppend(split[0])
            l_State.append(split[1])
            l_Country.append(None)
            
    else:
        l_City.append(split[0])
        l_State.append(split[1])
        l_Country.append(split[2])
        
for applicants in l_Applicants_Data:
    if data is not None:
        applicants=applicants.replace("Over 100 applicants","100+")
        split=applicants.split(" ")
        l_Applicants.append(split[0])
    else:
        l_Applicants.append(None)
    
    
for data in l_Data2:
    if data is not None:
        check = {
            "env": ["On-site", "Remote", "Hybrid"],
            "typ": ["Full-time", "Part-time", "Contract", "Temporary", "Volunteer", "Internship", "Other"],
            "exp": ["Associate", "Director", "Executive", "Mid-Senior-level", "Entry-level"]
        }
        replace = {"Mid-Senior level": "Mid-Senior-level", "Entry level": "Entry-level"}
        for old, new in replace.items():
            data = data.replace(old, new)
        split = data.split(" ")
        if len(split) <= 3:
            environment_found = False
            type_found = False
            experience_level_found = False
            for item in split:
                for key, value in check.items():
                    if item in value:
                        if key == "env":
                            l_Environment.append(item)
                            environment_found = True
                        elif key == "typ":
                            l_Type.append(item)
                            type_found = True
                        elif key == "exp":
                            l_Experience_Level.append(item)
                            experience_level_found = True
            
            if not environment_found:
                l_Environment.append(None)
            if not type_found:
                l_Type.append(None)
            if not experience_level_found:
                l_Experience_Level.append(None)
        else:
            l_Environment.append(None)
            l_Type.append(None)
            l_Experience_Level.append(None)
    else:
        l_Environment.append(None)
        l_Type.append(None)
        l_Experience_Level.append(None)
    
            
for data in l_Company_Data:
    if data is not None:
        split1= re.split(r'\s(?=\d)', data)
        if(len(split1)==3):
            l_Industry.append(split1[0])
            split2=split1[1].split(" ")
            l_Employees.append(split2[0])
            split3=split1[2].split(" ")
            l_Employees_On_Linkedin.append(split3[0])
        else:
            l_Industry.append(None)
            l_Employees.append(None)
            l_Employees_On_Linkedin.append(None)

    else:
        l_Industry.append(None)
        l_Employees.append(None)
        l_Employees_On_Linkedin.append(None)
            
for data in l_Followers_Data:
    if data is not None:
        split=data.split(" ")
        l_Followers.append(split[0])
    else:
        l_Followers.append(None)
```
